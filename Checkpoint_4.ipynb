```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Making a dataframe

```python
carprice = pd.read_csv('CarPrice_Assignment.csv')
```

So, we are given a car dataset in which there are various specifications of different cars.

We're just exploring data visualization by making various comparisions with various plot types.

```python
x = carprice.price
y = carprice.peakrpm

z = zip(x,y)

res = sorted(z, key = lambda x: x[0])

x_val = [m[0] for m in res]
y_val = [m[1] for m in res]

plt.plot(x_val, y_val)
plt.title("Peak rpm->price plot")
plt.xlabel("Price of car")
plt.ylabel("Peak rpm")
```
